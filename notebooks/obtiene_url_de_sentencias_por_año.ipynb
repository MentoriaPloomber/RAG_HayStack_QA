```python
def get_url_by_year(year_s_list= [2022]):

  """
  Input: Se ingresa una lista de años desde 1992 hasta 2022, por defecto esta el año 2022
  Return: se obtiene una lista de url de sentencias de tutela de la Corte Constitucional de Colombia

  Ejemplo:
  sentencias_2000 = get_url_by_year([2000])
  len(sentencias_2000)
  >>>3516

  sentencias_2005_2000 = get_url_by_year([2000, 2005])
  len(sentencias_2005_2000)
  >>>6178
  """
  #Convierte la lista de años en texto
  year_s_list = [str(year) for year in year_s_list]
  #Se inicializa la variable url donde se depositaran los link que se retornaran
  url = []
  #Diccionario de años con el numero total de sentencias de Tutela de cada año
  years_dict = {
      "1992": 615,
      "1993": 598,
      "1994": 580,
      "1995": 624,
      "1996": 717,
      "1997": 680,
      "1998": 805,
      "1999": 999,
      "2000": 1758,
      "2001": 1346,
      "2002": 1127,
      "2003": 1233,
      "2004": 1249,
      "2005": 1331,
      "2006": 1089,
      "2007": 1099,
      "2008": 1276,
      "2009": 974,
      "2010": 999,
      "2011": 981,
      "2012": 1097,
      "2013": 956,
      "2014": 977,
      "2015": 777,
      "2016": 737,
      "2017": 744,
      "2018": 501,
      "2019": 621,
      "2020": 536,
      "2021": 462,
      "2022": 471,
  }
  #Se itera por cada año ingresado como parametro
  for year in year_s_list:
    #si el año ingresado esta en las llaves del diccionario, se mide si la sentencia máxima supera o no los 1000 ejemplares
    #si supera los 1000 ejemplares se crean link del 1 al 999 y luego del 1000 hasta el máximo de numero de sentencias.
    if (year in years_dict.keys()) and (years_dict[year] < 1000):
      for step in range(1, years_dict[year] + 1):
        numero_text = '00'+ str(step)
        numero_text = numero_text[-3:]
        #print(numero_text)
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + numero_text + "-" + year[-2:] + ".htm")
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + numero_text + "A-" + year[-2:] + ".htm")
    else:
      for step in range(1, 1000):
        numero_text = '00'+ str(step)
        numero_text = numero_text[-3:]
        #print(numero_text)
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + numero_text + "-" + year[-2:] + ".htm")
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + numero_text + "A-" + year[-2:] + ".htm")
      for step in range(1000, years_dict[year] + 1):
        #print(numero_text)
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + str(step) + "-" + year[-2:] + ".htm")
        url.append("https://www.corteconstitucional.gov.co/relatoria/" + year + "/T-" + str(step) + "A-" + year[-2:] + ".htm")

    #Si la lista de años y la url esta vacia se retorna None
   # if len(list) == 0 and len(url) == 0:
    #  return None
  return url
```
