```python
import os
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
from datetime import datetime
import time

def download_and_convert_htm_links(links, txt_folder, sleep_time=30):
    if not os.path.exists(txt_folder):
        os.makedirs(txt_folder)

    total_links = len(links)
    time_estimate = total_links * sleep_time // 60  # Calcula el tiempo estimado en minutos

    confirmation = input(f"Tiempo aproximado de descarga: {time_estimate} minutos. ¿Desea continuar? (y/n): ")

    if confirmation.lower() != "y":
        print("Proceso de descarga cancelado.")
        return

    # Inicializa listas para rastrear los enlaces descargados y los fallidos
    downloaded_links = []
    failed_links = []

    # Registra la hora de inicio
    start_time = datetime.now()

    with tqdm(total=total_links, desc="Descargando archivos") as pbar:
        for link in links:
            response = requests.get(link, verify=False)

            if response.status_code == 200:
                if link.endswith(".htm"):
                    file_name = os.path.basename(link)
                    txt_name = file_name.replace(".htm", ".txt")
                    txt_path = os.path.join(txt_folder, txt_name)

                    # Procesa el contenido HTML utilizando BeautifulSoup
                    soup = BeautifulSoup(response.text, 'html.parser')
                    content_text = soup.get_text()

                    with open(txt_path, "w", encoding="utf-8") as txt_file:
                        txt_file.write(content_text)

                    downloaded_links.append(link)

            else:
                failed_links.append(link)

            pbar.update(1)
            time.sleep(sleep_time)

    # Registra la hora final
    end_time = datetime.now()

    # Genera un informe
    report_filename = "Informe_Descargas.txt"
    with open(report_filename, "w", encoding="utf-8") as report_file:
        report_file.write(f"Hora de inicio: {start_time}\n")
        report_file.write(f"Hora de finalización: {end_time}\n")
        report_file.write(f"Enlaces descargados ({len(downloaded_links)}):\n")
        for link in downloaded_links:
            report_file.write(f"  - {link}\n")
        report_file.write(f"Enlaces fallidos ({len(failed_links)}):\n")
        for link in failed_links:
            report_file.write(f"  - {link}\n")

    print(f"Informe generado: {report_filename}")

```
